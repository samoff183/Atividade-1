1. O Guerreiro (10 pontos): Precisamos representar um Guerreiro. Pense nas características essenciais que ele deve ter (como nome, pontos de vida, força de ataque) e crie uma classe para representá-lo.


```python
class Guerreiro:
    def __init__(self, nome, pontos_vida, forca_ataque):
        self.nome = nome
        self.pontos_vida = pontos_vida
        self.forca_ataque = forca_ataque
        self.arma = None

    def __str__(self):
        arma_info = f", Arma: {self.arma.nome} ({self.arma.poder_dano} dano)" if self.arma else ""
        return f"Guerreiro: {self.nome}, Vida: {self.pontos_vida}, Ataque: {self.forca_ataque}{arma_info}"

    def equipar_arma(self, arma):
        self.arma = arma
        print(f"{self.nome} equipou {self.arma.nome}.")

    def atacar(self):
        dano_total = self.forca_ataque
        if self.arma:
            dano_total += self.arma.poder_dano
            print(f"{self.nome} ataca com {self.arma.nome} causando {dano_total} de dano!")
        else:
            print(f"{self.nome} ataca com um soco feroz causando {dano_total} de dano!")

guerreiro = Guerreiro("He Man", 100, 70)
print(guerreiro)
```

11. Evitando Repetição (15 pontos): Guerreiros, Magos e Arqueiros possuem vários atributos em comum. Como você pode criar uma estrutura base Personagem para evitar a repetição de código entre essas classes?


### 11. Evitando Repetição com Herança

Vamos criar uma classe base `Personagem` e fazer com que `Guerreiro`, `Mago` e `Arqueiro` herdem dela, eliminando a repetição de código para atributos como `nome`, `pontos_vida`, `forca_ataque` e métodos como `__str__` e `atacar` (com a possibilidade de sobrescrevê-los).


```python
class Personagem:
    def __init__(self, nome, pontos_vida, forca_ataque):
        self.nome = nome
        self._pontos_vida = pontos_vida
        self.vida_maxima = pontos_vida
        self.forca_ataque = forca_ataque
        self.inventario = Inventario()

    @property
    def pontos_vida(self):
        return self._pontos_vida

    @pontos_vida.setter
    def pontos_vida(self, valor):
        if valor < 0:
            print(f"Vida de {self.nome} não pode ser negativa! Ajustando para 0.")
            self._pontos_vida = 0
        elif valor > self.vida_maxima:
            print(f"Vida de {self.nome} ultrapassou a vida máxima! Ajustando para {self.vida_maxima}.")
            self._pontos_vida = self.vida_maxima
        else:
            self._pontos_vida = valor

    def __str__(self):
        return f"Personagem: {self.nome}, Vida: {self.pontos_vida}/{self.vida_maxima}, Ataque: {self.forca_ataque}"

    def atacar(self, alvo):
        print(f"{self.nome} realiza um ataque básico em {alvo.nome}!")
        alvo.receber_dano(self.forca_ataque)

    def receber_dano(self, dano):
        print(f"{self.nome} recebeu {dano} de dano!")
        self.pontos_vida -= dano
        print(f"Vida atual de {self.nome}: {self.pontos_vida}")
        if self.pontos_vida == 0:
            print(f"{self.nome} foi derrotado!")

    def usar_item(self, item):
        if item in self.inventario.itens:
            if isinstance(item, Pocao):
                print(f"{self.nome} usou {item.nome} e curou {item.valor_cura} pontos de vida!")
                self.pontos_vida += item.valor_cura
                self.inventario.remover_item(item)
            else:
                print(f"{self.nome} não sabe como usar {item.nome}.")
        else:
            print(f"{item.nome} não está no inventário de {self.nome}.")

    def esta_vivo(self):
        return self.pontos_vida > 0


class Guerreiro(Personagem):
    def __init__(self, nome, pontos_vida, forca_ataque):
        super().__init__(nome, pontos_vida, forca_ataque)
        self.arma = None

    def __str__(self):
        arma_info = f", Arma: {self.arma.nome} ({self.arma.poder_dano} dano)" if self.arma else ""
        return f"Guerreiro: {self.nome}, Vida: {self.pontos_vida}/{self.vida_maxima}, Ataque: {self.forca_ataque}{arma_info}"

    def equipar_arma(self, arma):
        self.arma = arma
        print(f"{self.nome} equipou {self.arma.nome}.")

    def atacar(self, alvo):
        dano_total = self.forca_ataque
        if self.arma:
            dano_total += self.arma.poder_dano
            print(f"{self.nome} ataca com {self.arma.nome} causando {dano_total} de dano em {alvo.nome}!")
        else:
            print(f"{self.nome} ataca com um soco feroz causando {dano_total} de dano em {alvo.nome}!")
        alvo.receber_dano(dano_total)


class Mago(Personagem):
    def __init__(self, nome, pontos_vida, forca_ataque, poder_magico):
        super().__init__(nome, pontos_vida, forca_ataque)
        self.poder_magico = poder_magico

    def __str__(self):
        return f"Mago: {self.nome}, Vida: {self.pontos_vida}/{self.vida_maxima}, Ataque: {self.forca_ataque}, Magia: {self.poder_magico}"

    def atacar(self, alvo):
        dano_total = self.poder_magico # Mago usa poder mágico como dano
        print(f"{self.nome} conjura um feitiço poderoso em {alvo.nome} causando {dano_total} de dano mágico!")
        alvo.receber_dano(dano_total)


class Arqueiro(Personagem):
    def __init__(self, nome, pontos_vida, forca_ataque, precisao):
        super().__init__(nome, pontos_vida, forca_ataque)
        self.precisao = precisao

    def __str__(self):
        return f"Arqueiro: {self.nome}, Vida: {self.pontos_vida}/{self.vida_maxima}, Ataque: {self.forca_ataque}, Precisão: {self.precisao}"

    def atacar(self, alvo):
        dano_total = self.forca_ataque
        print(f"{self.nome} dispara uma flecha certeira em {alvo.nome} com {self.precisao}% de precisão, causando {dano_total} de dano!")
        alvo.receber_dano(dano_total)


class Monstro(Personagem):
    def __init__(self, nome, pontos_vida, forca_ataque):
        super().__init__(nome, pontos_vida, forca_ataque)

    def __str__(self):
        return f"Monstro: {self.nome}, Vida: {self.pontos_vida}/{self.vida_maxima}, Ataque: {self.forca_ataque}"

    def atacar(self, alvo):
        dano_total = self.forca_ataque
        print(f"{self.nome} investe com fúria contra {alvo.nome} causando {dano_total} de dano!")
        alvo.receber_dano(dano_total)

    @classmethod
    def criar_goblin_padrao(cls, nome="Goblin Padrão"):
        return cls(nome, 50, 20)
```

15. O Inventário (15 pontos): Heróis colecionam muitos itens. Modele uma classe Inventário que possa guardar uma coleção de itens (como armas e poções).


```python
class Inventario:
    def __init__(self):
        self.itens = []

    def adicionar_item(self, item):
        self.itens.append(item)
        print(f"{item.nome} adicionado ao inventário.")

    def remover_item(self, item):
        if item in self.itens:
            self.itens.remove(item)
            print(f"{item.nome} removido do inventário.")
        else:
            print(f"{item.nome} não encontrado no inventário.")

    def listar_itens(self):
        if not self.itens:
            print("Inventário vazio.")
            return
        print("Itens no inventário:")
        for item in self.itens:
            print(f"- {item}")
```

### 21. Molde de Habilidade com Classes Abstratas

Vamos usar o módulo `abc` do Python para criar uma `Habilidade` como uma classe abstrata, garantindo que todas as habilidades concretas implementem um método `usar()`.

```python
from abc import ABC, abstractmethod

class Habilidade(ABC):
    def __init__(self, nome, custo_mana=0):
        self.nome = nome
        self.custo_mana = custo_mana

    @abstractmethod
    def usar(self, atacante, alvo):
        pass

    def __str__(self):
        return f"Habilidade: {self.nome} (Custo de Mana: {self.custo_mana})"


class AtaqueEspada(Habilidade):
    def __init__(self):
        super().__init__("Ataque do Castelo de Grayskull", custo_mana=10)

    def usar(self, atacante, alvo):
        dano = atacante.forca_ataque * 1.5
        print(f"{atacante.nome} usa {self.nome} em {alvo.nome}, causando {dano:.0f} de dano extra!")
        alvo.receber_dano(dano)


class MAGIA(Habilidade):
    def __init__(self):
        super().__init__("Avada Kedavra", custo_mana=20)

    def usar(self, atacante, alvo):
        dano = atacante.poder_magico * 1.2
        print(f"{atacante.nome} lança uma {self.nome} em {alvo.nome}, causando {dano:.0f} de dano mágico!")
        alvo.receber_dano(dano)


guerreiro = Guerreiro("He Man", 100, 70)
mago = Mago("Voldermmort", 50, 60, 70)
alvo_teste = Monstro("Tifão", 150, 25)

print("--- Demonstração de Habilidades ---")
print(guerreiro)
print(mago)
print(alvo_teste)

print(f"\n{guerreiro.nome} tem {guerreiro.pontos_vida} de vida.")
print(f"{alvo_teste.nome} tem {alvo_teste.pontos_vida} de vida.")
ataque_forte = AtaqueEspada()
ataque_forte.usar(guerreiro, alvo_teste)
print(f"{alvo_teste.nome} agora tem {alvo_teste.pontos_vida} de vida.")

print(f"\n{mago.nome} tem {mago.pontos_vida} de vida.")
print(f"{alvo_teste.nome} tem {alvo_teste.pontos_vida} de vida.")
bola_de_fogo = MAGIA()
bola_de_fogo.usar(mago, alvo_teste)
print(f"{alvo_teste.nome} agora tem {alvo_teste.pontos_vida} de vida.")
```
