1. O Guerreiro (10 pontos): Precisamos representar um Guerreiro. Pense nas características essenciais que ele deve ter (como nome, pontos de vida, força de ataque) e crie uma classe para representá-lo.


```python
class Guerreiro:
    def __init__(self, nome, pontos_vida, forca_ataque):
        self.nome = nome
        self.pontos_vida = pontos_vida
        self.forca_ataque = forca_ataque
        self.arma = None

    def __str__(self):
        arma_info = f", Arma: {self.arma.nome} ({self.arma.poder_dano} dano)" if self.arma else ""
        return f"Guerreiro: {self.nome}, Vida: {self.pontos_vida}, Ataque: {self.forca_ataque}{arma_info}"

    def equipar_arma(self, arma):
        self.arma = arma
        print(f"{self.nome} equipou {self.arma.nome}.")

    def atacar(self):
        dano_total = self.forca_ataque
        if self.arma:
            dano_total += self.arma.poder_dano
            print(f"{self.nome} ataca com {self.arma.nome} causando {dano_total} de dano!")
        else:
            print(f"{self.nome} ataca com um soco feroz causando {dano_total} de dano!")

guerreiro = Guerreiro("He Man", 100, 70)
print(guerreiro)
```

### 11. Evitando Repetição com Herança

Vamos criar uma classe base `Personagem` e fazer com que `Guerreiro`, `Mago` e `Arqueiro` herdem dela, eliminando a repetição de código para atributos como `nome`, `pontos_vida`, `forca_ataque` e métodos como `__str__` e `atacar` (com a possibilidade de sobrescrevê-los).


```python
from abc import ABC, abstractmethod
import random
class Dado:
    def __init__(self, lados):
        if not isinstance(lados, int) or lados < 1:
            raise ValueError("O número de lados do dado deve ser um inteiro positivo.")
        self.lados = lados

    def rolar(self):
        return random.randint(1, self.lados)

    def __str__(self):
        return f"Dado de {self.lados} lados"


class Personagem:
    def __init__(self, nome, pontos_vida, forca_ataque):
        self.nome = nome
        self._pontos_vida = pontos_vida
        self.vida_maxima = pontos_vida
        self.forca_ataque = forca_ataque
        self.inventario = Inventario()
        self.habilidades = []
        self.dado = Dado(6)
    @property
    def pontos_vida(self):
        return self._pontos_vida

    @pontos_vida.setter
    def pontos_vida(self, valor):
        if valor < 0:
            print(f"Vida de {self.nome} não pode ser negativa! Ajustando para 0.")
            self._pontos_vida = 0
        elif valor > self.vida_maxima:
            print(f"Vida de {self.nome} ultrapassou a vida máxima! Ajustando para {self.vida_maxima}.")
            self._pontos_vida = self.vida_maxima
        else:
            self._pontos_vida = valor

    def __str__(self):
        return f"Personagem: {self.nome}, Vida: {self.pontos_vida}/{self.vida_maxima}, Ataque: {self.forca_ataque}"

    def atacar(self, alvo):
        dano_base = self.forca_ataque
        dano_variacao = self.dado.rolar() - (self.dado.lados // 2 + 1)
        dano_total = max(0, dano_base + dano_variacao)
        print(f"{self.nome} realiza um ataque básico em {alvo.nome}! (Dano base: {dano_base}, Variação: {dano_variacao}, Total: {dano_total:.0f})")
        alvo.receber_dano(dano_total)

    def receber_dano(self, dano):
        print(f"{self.nome} recebeu {dano:.0f} de dano!")
        self.pontos_vida -= dano
        print(f"Vida atual de {self.nome}: {self.pontos_vida}")
        if self.pontos_vida == 0:
            print(f"{self.nome} foi derrotado!")

    def usar_item(self, item):
        if item in self.inventario.itens:
            if isinstance(item, Pocao):
                print(f"{self.nome} usou {item.nome} e curou {item.valor_cura} pontos de vida!")
                self.pontos_vida += item.valor_cura
                self.inventario.remover_item(item)
            else:
                print(f"{self.nome} não sabe como usar {item.nome}.")
        else:
            print(f"{item.nome} não está no inventário de {self.nome}.")

    def esta_vivo(self):
        return self.pontos_vida > 0

    def aprender_habilidade(self, habilidade):
        self.habilidades.append(habilidade)
        print(f"{self.nome} aprendeu a habilidade: {habilidade.nome}.")

    def usar_habilidade(self, nome_habilidade, alvo):
        for habilidade in self.habilidades:
            if habilidade.nome == nome_habilidade:
                print(f"{self.nome} usa sua habilidade '{habilidade.nome}'!")
                habilidade.usar(self, alvo)
                return
        print(f"{self.nome} não conhece a habilidade '{nome_habilidade}'.")


class Guerreiro(Personagem):
    def __init__(self, nome, pontos_vida, forca_ataque):
        super().__init__(nome, pontos_vida, forca_ataque)
        self.arma = None

    def __str__(self):
        arma_info = f", Arma: {self.arma.nome} ({self.arma.poder_dano} dano)" if self.arma else ""
        return f"Guerreiro: {self.nome}, Vida: {self.pontos_vida}/{self.vida_maxima}, Ataque: {self.forca_ataque}{arma_info}"

    def equipar_arma(self, arma):
        self.arma = arma
        print(f"{self.nome} equipou {self.arma.nome}.")

    def atacar(self, alvo):
        dano_base = self.forca_ataque
        if self.arma:
            dano_base += self.arma.poder_dano

        dano_variacao = self.dado.rolar() - (self.dado.lados // 2 + 1)
        dano_total = max(0, dano_base + dano_variacao)

        if self.arma:
            print(f"{self.nome} ataca com {self.arma.nome} em {alvo.nome}! (Dano base: {dano_base - self.arma.poder_dano} + {self.arma.poder_dano} da arma, Variação: {dano_variacao}, Total: {dano_total:.0f})")
        else:
            print(f"{self.nome} ataca com um soco feroz em {alvo.nome}! (Dano base: {dano_base}, Variação: {dano_variacao}, Total: {dano_total:.0f})")
        alvo.receber_dano(dano_total)


class Mago(Personagem):
    def __init__(self, nome, pontos_vida, forca_ataque, poder_magico):
        super().__init__(nome, pontos_vida, forca_ataque)
        self.poder_magico = poder_magico

    def __str__(self):
        return f"Mago: {self.nome}, Vida: {self.pontos_vida}/{self.vida_maxima}, Ataque: {self.forca_ataque}, Magia: {self.poder_magico}"

    def atacar(self, alvo):
        dano_base = self.poder_magico
        dano_variacao = self.dado.rolar() - (self.dado.lados // 2 + 1)
        dano_total = max(0, dano_base + dano_variacao)

        print(f"{self.nome} conjura um feitiço poderoso em {alvo.nome}! (Dano base: {dano_base}, Variação: {dano_variacao}, Total: {dano_total:.0f})")
        alvo.receber_dano(dano_total)


class Arqueiro(Personagem):
    def __init__(self, nome, pontos_vida, forca_ataque, precisao):
        super().__init__(nome, pontos_vida, forca_ataque)
        self.precisao = precisao

    def __str__(self):
        return f"Arqueiro: {self.nome}, Vida: {self.pontos_vida}/{self.vida_maxima}, Ataque: {self.forca_ataque}, Precisão: {self.precisao}"

    def atacar(self, alvo):
        dano_base = self.forca_ataque
        dano_variacao = self.dado.rolar() - (self.dado.lados // 2 + 1)
        dano_total = max(0, dano_base + dano_variacao)

        print(f"{self.nome} dispara uma flecha certeira em {alvo.nome}! (Dano base: {dano_base}, Variação: {dano_variacao}, Total: {dano_total:.0f})")
        alvo.receber_dano(dano_total)


class Monstro(Personagem):
    def __init__(self, nome, pontos_vida, forca_ataque):
        super().__init__(nome, pontos_vida, forca_ataque)

    def __str__(self):
        return f"Monstro: {self.nome}, Vida: {self.pontos_vida}/{self.vida_maxima}, Ataque: {self.forca_ataque}"

    def atacar(self, alvo):
        dano_base = self.forca_ataque
        dano_variacao = self.dado.rolar() - (self.dado.lados // 2 + 1)
        dano_total = max(0, dano_base + dano_variacao)

        print(f"{self.nome} investe com fúria contra {alvo.nome}! (Dano base: {dano_base}, Variação: {dano_variacao}, Total: {dano_total:.0f})")
        alvo.receber_dano(dano_total)

    @classmethod
    def criar_goblin_padrao(cls, nome="Goblin Padrão"):
        return cls(nome, 50, 20)


# Inventario class definition (assuming it's available or defined earlier in the notebook)
class Inventario:
    def __init__(self):
        self.itens = []

    def adicionar_item(self, item):
        self.itens.append(item)
        print(f"{item.nome} adicionado ao inventário.")

    def remover_item(self, item):
        if item in self.itens:
            self.itens.remove(item)
            print(f"{item.nome} removido do inventário.")
        else:
            print(f"{item.nome} não encontrado no inventário.")

    def listar_itens(self):
        if not self.itens:
            print("Inventário vazio.")
            return
        print("Itens no inventário:")
        for item in self.itens:
            print(f"- {item}")

# Arma and Pocao classes (assuming they are available or defined earlier in the notebook)
class Arma:
    def __init__(self, nome, poder_dano):
        self.nome = nome
        self.poder_dano = poder_dano

    def __str__(self):
        return f"Arma: {self.nome}, Dano: {self.poder_dano}"

class Pocao:
    def __init__(self, nome, valor_cura):
        self.nome = nome
        self.valor_cura = valor_cura

    def __str__(self):
        return f"Poção: {self.nome}, Cura: {self.valor_cura}"
```

### 21. Molde de Habilidade com Classes Abstratas

Vamos usar o módulo `abc` do Python para criar uma `Habilidade` como uma classe abstrata, garantindo que todas as habilidades concretas implementem um método `usar()`.

```python
from abc import ABC, abstractmethod

class Habilidade(ABC):
    def __init__(self, nome, custo_mana=0):
        self.nome = nome
        self.custo_mana = custo_mana

    @abstractmethod
    def usar(self, atacante, alvo):
        pass

    def __str__(self):
        return f"Habilidade: {self.nome} (Custo de Mana: {self.custo_mana})"


class AtaqueEspada(Habilidade):
    def __init__(self):
        super().__init__("Ataque do Castelo de Grayskull", custo_mana=10)

    def usar(self, atacante, alvo):
        dano = atacante.forca_ataque * 1.5
        print(f"{atacante.nome} usa {self.nome} em {alvo.nome}, causando {dano:.0f} de dano extra!")
        alvo.receber_dano(dano)


class MAGIA(Habilidade):
    def __init__(self):
        super().__init__("Avada Kedavra", custo_mana=20)

    def usar(self, atacante, alvo):
        dano = atacante.poder_magico * 1.2
        print(f"{atacante.nome} lança uma {self.nome} em {alvo.nome}, causando {dano:.0f} de dano mágico!")
        alvo.receber_dano(dano)


guerreiro = Guerreiro("He Man", 100, 70)
mago = Mago("Voldermmort", 50, 60, 70)
alvo_teste = Monstro("Tifão", 150, 25)

print("--- Demonstração de Habilidades ---")
print(guerreiro)
print(mago)
print(alvo_teste)

print(f"\n{guerreiro.nome} tem {guerreiro.pontos_vida} de vida.")
print(f"{alvo_teste.nome} tem {alvo_teste.pontos_vida} de vida.")
ataque_forte = AtaqueEspada()
ataque_forte.usar(guerreiro, alvo_teste)
print(f"{alvo_teste.nome} agora tem {alvo_teste.pontos_vida} de vida.")

print(f"\n{mago.nome} tem {mago.pontos_vida} de vida.")
print(f"{alvo_teste.nome} tem {alvo_teste.pontos_vida} de vida.")
bola_de_fogo = MAGIA()
bola_de_fogo.usar(mago, alvo_teste)
print(f"{alvo_teste.nome} agora tem {alvo_teste.pontos_vida} de vida.")
```

```python
import random

class Dado:
    def __init__(self, lados):
        if not isinstance(lados, int) or lados < 1:
            raise ValueError("O número de lados do dado deve ser um inteiro positivo.")
        self.lados = lados

    def rolar(self):
        return random.randint(1, self.lados)

    def __str__(self):
        return f"Dado de {self.lados} lados"

dado_d6 = Dado(6)
dado_d20 = Dado(20)

print(f"Rolando um {dado_d6}: {dado_d6.rolar()}")
print(f"Rolando um {dado_d20}: {dado_d20.rolar()}")
print(f"Rolando outro {dado_d6}: {dado_d6.rolar()}")
```

### Subtask:
Definir uma nova classe `Time` (Team) que encapsulará uma lista de objetos `Personagem`. Esta classe deve ter métodos para adicionar membros, listar membros vivos e verificar se todos os membros da equipe foram derrotados.


**Reasoning**:
The subtask is to create a new class `Time` with specified methods. I will create a code block that defines the `Time` class, its constructor, and the required methods to manage team members.



```python
class Time:
    def __init__(self, nome):
        self.nome = nome
        self.membros = []

    def adicionar_membro(self, personagem):
        if isinstance(personagem, Personagem):
            self.membros.append(personagem)
            print(f"{personagem.nome} foi adicionado ao time {self.nome}.")
        else:
            print("Apenas instâncias de Personagem podem ser adicionadas ao time.")

    def get_membros_vivos(self):
        return [membro for membro in self.membros if membro.esta_vivo()]

    def todos_derrotados(self):
        return len(self.get_membros_vivos()) == 0

    def __str__(self):
        membros_vivos = self.get_membros_vivos()
        num_membros_vivos = len(membros_vivos)
        return f"Time: {self.nome} (Membros vivos: {num_membros_vivos}/{len(self.membros)})"

# Demonstração da classe Time
guerreiro1 = Guerreiro("Arthur", 150, 75)
mago1 = Mago("Gandalf", 100, 50, 120)
orc1 = Orc("Uruk-hai", 80, 35)

time_herois = Time("Guardiões da Luz")
time_herois.adicionar_membro(guerreiro1)
time_herois.adicionar_membro(mago1)

time_vilao = Time("Legião do Mal")
time_vilao.adicionar_membro(orc1)

print(time_herois)
print(time_vilao)

# Simular derrota de um membro para testar get_membros_vivos e todos_derrotados
print("\nSimulando combate...")
guerreiro1.receber_dano(200) # Derrota o guerreiro

print(f"\n{time_herois.nome}: {len(time_herois.get_membros_vivos())} membros vivos")
print(f"Todos os heróis derrotados? {time_herois.todos_derrotados()}")

mago1.receber_dano(1000) # Derrota o mago

print(f"\n{time_herois.nome}: {len(time_herois.get_membros_vivos())} membros vivos")
print(f"Todos os heróis derrotados? {time_herois.todos_derrotados()}")
```

## Adaptar Batalha para Equipes

### Subtask:
Alterar a classe `Batalha` para que receba dois objetos `Time` em seu construtor em vez de dois `Personagem`. A lógica do loop de combate (`iniciar_batalha`) precisará ser ajustada para gerenciar os turnos das equipes, onde cada membro vivo de uma equipe ataca um alvo aleatório da equipe adversária.


**Reasoning**:
The subtask requires modifying the `Batalha` class to manage team battles. This involves updating the constructor, implementing team turn-based combat, and selecting random targets from opposing teams. I will start by adding the necessary imports and modifying the `Batalha` class definition.



```python
import random

class Batalha:
    def __init__(self, team1, team2):
        if not isinstance(team1, Time) or not isinstance(team2, Time):
            raise ValueError("Ambos os combatentes devem ser instâncias de Time.")
        self.team1 = team1
        self.team2 = team2
        print(f"\n--- BATALHA DE EQUIPES: {self.team1.nome} vs. {self.team2.nome} ---")

    def iniciar_batalha(self):
        rodada = 1
        while not self.team1.todos_derrotados() and not self.team2.todos_derrotados():
            print(f"\n--- Rodada {rodada} ---")
            print(f"Status: {self.team1}")
            print(f"Status: {self.team2}")

            print(f"\n--- Turno de {self.team1.nome} ---")
            for atacante in self.team1.get_membros_vivos():
                if not self.team2.todos_derrotados():
                    alvo = random.choice(self.team2.get_membros_vivos())
                    atacante.atacar(alvo)
                else:
                    break

            if self.team2.todos_derrotados():
                break

            print(f"\n--- Turno de {self.team2.nome} ---")
            for atacante in self.team2.get_membros_vivos():
                if not self.team1.todos_derrotados():
                    alvo = random.choice(self.team1.get_membros_vivos())
                    atacante.atacar(alvo)
                else:
                    break

            rodada += 1

        self._declarar_vencedor()

    def _declarar_vencedor(self):
        print("\n--- FIM DA BATALHA ---")
        print(f"Status Final: {self.team1}")
        print(f"Status Final: {self.team2}")

        if self.team1.todos_derrotados() and self.team2.todos_derrotados():
            print("\n!!! AMBAS AS EQUIPES FORAM DERROTADAS !!!")
        elif self.team1.todos_derrotados():
            print(f"\n!!! {self.team2.nome} é a equipe VENCEDORA !!!")
        else:
            print(f"\n!!! {self.team1.nome} é a equipe VENCEDORA !!!")

```
